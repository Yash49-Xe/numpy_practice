```python
#18
mat_5x5 = np.zeros((5,5))
# mat_5x5

c = 1
for i in range(5):
  for j in range(5):
    if i - j == 1:
      mat_5x5[i][j] = c
      c += 1
mat_5x5
```

```python
#19
mat_8x8 = np.zeros((8,8), dtype = np.int64)
for i in range(8):
  c = i
  for j in range(8):
    if c % 2 == 0:
      mat_8x8[i][j] = 1
    c += 1
mat_8x8
```
